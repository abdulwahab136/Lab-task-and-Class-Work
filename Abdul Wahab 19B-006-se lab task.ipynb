```python
#program to solve quardatic equations
from math import sqrt

a = eval(input("Enter value of a :"))
b = eval(input("Enter value of b :"))
c = eval(input("Enter value of c :"))


if 2*a == 0:
    print("the equation cannot be solved as there is a zero division")
else:
    x1= (-b+sqrt(b**2-4*a*c))/2*a
    x2= (-b-sqrt(b**2-4*a*c))/2*a
    print("First Solution is : ",x1)
    print("Second Solution is : ",x2)
```

```python
#program to check is string is palindrome or not

word = str(input("Enter String you want to know if it's palindrome :"))
rword= ''.join(reversed(word))


if word == rword:
    print("The word ",word," is palindrome")
else:
    print("Sorry ",word," is not palindrome")
```

```python
#program for marksheet

name = input(print("Enter your name : "))
fname = input(print("Enter your father name : "))
rno = input(print("Enter your roll number : "))
year = input(print("Enter the year you did your Intermediate :"))
total = 0

slist = ['','','','','']
mlist = [0,0,0,0,0]

for i in range(0,5):
    slist[i] = input(print("Enter Subject : "))
    mlist[i] = int(input(print("Enter Marks Obtained : ")))
    total = total + mlist[i]


print("\t\t\t\tIntermidiate Examination Marksheet ",year)
print("Name : ",name)
print("Father Name : ",fname)
print("Roll Number : ",rno)
print("Subject \t\t Total Marks \t\t Marks Obtained ")

for i in range(0,5):
    print(slist[i] ,"\t\t\t", "100" , "\t\t\t",mlist[i])
    
print("Total Marks : 500")
print("Marks Obtained :",total)
percentage = float((total/500)*100) 
print("Percentage : ",percentage)
if percentage > 80:
    print("Grade A+")
else:
    if percentage > 70:
        print("Grade A")
    else:
        if percentage > 60:
            print("Grade B")
        else:
            if percentage > 50:
                print("Grade C")
            else:
                if percentage > 33:
                    print("Grade D")
                else:
                    print("Grade F")


    


    
```
