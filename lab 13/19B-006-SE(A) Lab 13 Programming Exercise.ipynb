```python
#Task 4
vowel = 0
infile = open("RE.txt",'r+')
content = infile.read()
words = content.split()
setwords = set(words)
newlist = []

#total vovels
for word in words:
    for letter in word:
        if letter in 'AEIOUaeiou':
            vowel = vowel + 1
print("Total Number of Vowels : ",vowel)
print("Total Number of words  : ",len(words))
print("\nDuplicate Words")
print("-----------------")
for word in setwords:
    word.casefold()
    dword = words.count(word)
    if dword > 1:
        print("Word {0} repeated {1} times".format(word,dword))
        
for word in words:
    newlist.append(word.upper())

print(newlist)
```
